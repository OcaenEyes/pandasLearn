### Pandas可视化

- 折线图
- 柱状图
- 直方图
- 箱线图
- 区域图
- 散点图
- 饼图
- 六边形容器图

```python
# 导入相关库
import numpy as np
import pandas as pd

import matplotlib.pyplot as pt
import matplotlib

%matplotlib inline

np.random.seed(100)
```

```python
df = pd.DataFrame(np.random.randint(-10,10,(10,3)), index = pd.date_range("1/1/2019",periods = 10), columns = list("ABC"))
```

```python
df = df.cumsum()
```

```python
df.head()
```

#### 折线图

- 默认情况，参数 kind="line"
- 将索引作为 x轴， 列作为 y轴

```python
df.plot()
```

```python
df.plot(kind="line")
```

- 指定x轴 ，指定y轴

```python
df.plot(x="A",y="C")
```

#### 柱状图

- kind 设置为 bar ／  barh

```python
# 索引为X轴， 列为Y轴
df.plot(kind="bar")
```

```python
# 索引为Y轴， 列为X轴
df.plot(kind="barh")
```

- 指定X轴，Y轴

```python
df.plot(kind="bar",x= "A", y = ["B","C"])
```

#### 直方图

- kind ="hist"

```python
df.plot(kind="hist")
```

```python
# 设置bins 参数 改变bin的大小
df.plot(kind="hist",bins = 5)
```

#### 箱线图

- 箱线图展示 分位数；具体包括上四分位数、下四分位数、中位数、以及上下5%的极值
- kind ="box"

```python
df.plot(kind="box")
```

#### 区域图

- kind ="area"
- 默认状态，区域图是堆积的，要生成堆积图，每列必须为全正值、或全负值

```python
df.abs().plot(kind="area")
```

```python
# 若需要生成不堆积的区域图，设置参数 stacked= False
df.abs().plot(kind="area",stacked=False)
```

#### 散点图

- kind ="scatter"
- 指定X 和Y
- 通过散点图，探索变量之间的关系

```python
df.plot(kind="scatter",x="A", y="B")
```

```python
# 设置参数C ,为每个点设置提供颜色
df.plot(kind="scatter",x="A",y="B",c="C")
```

```python
# 单个轴上绘制多列，需指定ax
ax = df.plot(kind="scatter",x="A",y="B",color="blue")
```

```python
df.plot(kind="scatter",x="C",y="B",color="green",ax = ax)
```

#### 饼图

```python
a = df.A[:5]
```

```python
a.abs().plot.pie(subplots= True,figsize = (4,4))
```

```python
# 自动计算出比例
a.abs().plot.pie(subplots=True,figsize=(4,4),autopct = "%.2f")
```

#### 六边形容器图

- 数据过于密集，无法单独绘制每个点，考虑Hexbin

```python
df = pd.DataFrame(np.random.randn(1000,2),columns = ["A","B"])
```

```python
df["B"] = df ["B"] + np.arange(1000)
```

```python
df.plot(kind="hexbin", x="A", y="B",gridsize = 10)
```

```python
# gridsize控制x轴，六边形的数量，默认为100
df.plot(kind="hexbin", x="A", y="B",gridsize = 20)
```
