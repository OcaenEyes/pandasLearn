```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path1 = './dataset/exercise_data/train.csv'
```

```python
titanic = pd.read_csv(path1)
```

```python
titanic
```

```python
# å°†PassengerIdè®¾ç½®ä¸ºç´¢å¼•
titanic.set_index('PassengerId')
```

```python
# ç»˜åˆ¶ç”·å¥³ä¹˜å®¢æ¯”ä¾‹çš„æ‰‡å½¢å›¾
males = (titanic['Sex'] == 'male').sum()
# males
females = (titanic['Sex'] == 'female').sum()
# females
proportions = [males,females]
proportions
```

```python
plt.pie(
    proportions,
    labels=['Males','Females'],
    shadow=False,
    colors=['red','green'],
    explode=(0.15,0),
    startangle = 90,
    autopct = '%1.1f%%'
)
plt.axis('equal')
plt.title('Sex Propertion')
plt.tight_layout()
plt.show()
```

```python
# ç»˜åˆ¶å±•ç¤ºèˆ¹ç¥¨Fare ä¸Žä¹˜å®¢å¹´é¾„ã€æ€§åˆ«çš„æ•£ç‚¹å›¾
lm = sns.lmplot(x='Age' , y ='Fare', data=titanic, hue='Sex', fit_reg=False)

# è®¾ç½®title
lm.set(title='Fare * Age')

axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```
