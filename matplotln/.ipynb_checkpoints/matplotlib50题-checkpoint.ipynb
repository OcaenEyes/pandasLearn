```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 基本图表
# plot绘制x=(0,10)的sin图像
x = np.linspace(0,10,num=30)
plt.plot(x,np.sin(x))
```

```python
# 用点加线的方式 画出x=(0,10)的 sin图像
plt.plot(x,np.sin(x), '-o')
```

```python
# 用scatter 方法绘制x=(0,10)的 sin图像
plt.scatter(x,np.sin(x))
```

```python
# 使用饼图的面积及颜色绘制一组四维数据
# 伪随机数生成器
rng = np.random.RandomState(0)

x= rng.randn(100)
y= rng.randn(100)
colors = rng.rand(100)
sizes =  1000 * rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # 展示色阶
```

```python
# 绘制一组误差为+- 0.8的数据误差条图
x= np.linspace(0,10,num=50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy, fmt='.k')
```

```python
# 绘制一个柱形图
x= [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]
label = ['A','B','C','D','E','F','G','H']
plt.bar(x,y,tick_label = label)
```

```python
# 绘制一个水平方向的柱形图
# 绘制一个柱形图
x= [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]
label = ['A','B','C','D','E','F','G','H']
plt.barh(x,y,tick_label = label)
```

```python
# 绘制1000哥随机值的直方图
data = np.random.randn(1000)
plt.hist(data)
```

```python
# 设置直方图 分30个bins ，并设置为频率分布
plt.hist(data, bins=30, histtype='stepfilled', density=True)
```

```python
# 在一张图内绘制3组不同的直方图， 并设置透明度
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)

kwargs = dict(alpha=0.3 , bins=40 , density = True)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.show()
```

```python
# 绘制一张二维直方图
mean =[0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,10000).T
plt.hist2d(x,y,bins=30)
plt.show()
```

```python
# 绘制一张 设置网格大小为30 的六角直方图
plt.hexbin(x,y,gridsize=30)
```

```python
# 设x,y轴标签 
# 设置显示汉字
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y,label='sin(x)')
plt.xlabel('x轴数值')
plt.ylabel('y轴数值')
```

```python
# 设置图表标题 “三角函数”
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y,label='sin(x)')
plt.title('三角函数')
plt.xlabel('x轴')
plt.ylabel('y轴')
```

```python
# 显示网格
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y,label='sin(x)')
plt.title('三角函数')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.grid()
```

```python
# 绘制平行于x轴 y=0.8的水平参考线
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y,label='sin(x)')
plt.title('三角函数')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.grid()
plt.axhline(y=0.8, ls='--', c='r')
```

```python
# 绘制垂直于x轴 x>4 and x<6的 ， 以及y轴  -0.2<y<0.2的参考区域
x= np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.axvspan(xmin=4,xmax=6, facecolor='r',alpha=0.3) #垂直x轴
plt.axhspan(ymin=-0.2 ,ymax=0.2, facecolor='y', alpha=0.3) # 垂直y轴
plt.show()
```

```python
# 添加文字注释 sin(x)
x = np.linspace(0.05,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.text(3.2,0,'sin(x)',weight='bold',color='r')
```

```python
# 用箭头标出第一个峰值
x =np.linspace(0.05,10,100)
y= np.sin(x)
plt.plot(x,y)
plt.annotate('maximum',
             xy=(np.pi/2,1),
             xytext=(np.pi/2+1,1),
             weight='bold',
             color='r',
             arrowprops= dict(arrowstyle='->',
                              connectionstyle='arc3',
                              color='r'
                             )
            )
```

## 自定义图例

```python
# 在一张图里绘制sin, cos图形，并展示图例
x= np.linspace(0,10,100)
fig,ax = plt.subplots()
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),'--',label='cos')
ax.legend()
```

```python
# 调整图例左上角展示，且不显示边框
ax.legend(loc='upper left', frameon=False)
fig
```

```python
# 调整图例再画面下方居中显示，分成2列
ax.legend(frameon=False, loc='lower center' ,ncol=2)
fig
```

```python
# 绘制图像sin(x) ,sin(x +π/2) ， sin(x+π) ，只显示前2者的图例

y= np.sin(x[:,np.newaxis] + np.pi *np.arange(0,2,0.5))
lines = plt.plot(x,y)
# lines 是plt.Line2D 类型的实例的列表
plt.legend(lines[:2],['first','second'])
```

```python
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True)
```

```python
# 将图例分不同的区域展示
fig,ax = plt.subplots()
lines=[]
styles= ['-','--','-.',':']
x = np.linspace(0,10,1000)
for i in range(4):
    lines = lines + ax.plot(x,np.sin(x-i*np.pi /2),styles[i],color='black')
ax.axis('equal')

# 设置第一组标签
ax.legend(lines[:2],
          ['line A','line B'],
          loc='upper right',
          frameon = False
         )

# 设置第二组标签
from matplotlib.legend import Legend
leg = Legend(ax,lines[2:],
             ['line C','line D'],
             loc='lower right',
             frameon =False
            )
ax.add_artist(leg)
```

## 自定义色阶

```python
# 展示色阶
x = np.linspace(0,10,1000)
I = np.sin(x) * np.cos(x[:,np.newaxis])

plt.imshow(I)
plt.colorbar()
```

```python
# 改变配色为 'gray'
plt.imshow(I,cmap='gray')
```

```python
# 将色阶分成 6个离散值显示
plt.imshow(I,cmap=plt.cm.get_cmap('Blues',6))
plt.colorbar()
plt.clim(-1,1)
```

## 多子图

```python
# 在一个10*10的画布中，(0.65,0.65)的位置创建一个0.2*0.2的子图
ax1 = plt.axes()
ax2 = plt.axes([0.65,0.65,0.2,0.2])
```

```python
# 在两个子图中 显示 sin(x) 和cos(x)的图像
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.5,0.8,0.4], ylim=(-1.2,1.2))
ax2 = fig.add_axes([0.1,0.1,0.8,0.4],ylim=(-1.2,1.2))
x = np.linspace(0,10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

```python
# 用for创建6个子图， 并在图中标识对应的子图坐标
for i in range(1,7):
    print(i)
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize=18,ha='center')
```

```python
# 设置同行和列 共享x,y轴
fig,ax = plt.subplots(2,3,sharex='col',sharey='row')
```

```python
# 用[]的方式取出 每个子图， 并添加子图坐标文字
for i in range(2):
    for j in range(3):
        print(i,j)
        ax[i,j].text(0.5,0.5,str((i,j)),fontsize=18, ha='center')
fig
```

```python
# 组合绘制大小不同的子图
grid = plt.GridSpec(2,3,wspace=0.4,hspace=0.3)
plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])
plt.subplot(grid[1,:2])
plt.subplot(grid[1,2])
```
