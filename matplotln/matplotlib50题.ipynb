```python
# 使用饼图的面积及颜色绘制一组四维数据
# 伪随机数生成器
rng = np.random.RandomState(0)

x= rng.randn(100)
y= rng.randn(100)
colors = rng.rand(100)
sizes =  1000 * rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # 展示色阶
```

```python
# 绘制一组误差为+- 0.8的数据误差条图
x= np.linspace(0,10,num=50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy, fmt='.k')
```

## 自定义图例

```python
# 调整图例左上角展示，且不显示边框
ax.legend(loc='upper left', frameon=False)
fig
```

```python
# 调整图例再画面下方居中显示，分成2列
ax.legend(frameon=False, loc='lower center' ,ncol=2)
fig
```

```python
# 绘制图像sin(x) ,sin(x +π/2) ， sin(x+π) ，只显示前2者的图例

y= np.sin(x[:,np.newaxis] + np.pi *np.arange(0,2,0.5))
lines = plt.plot(x,y)
# lines 是plt.Line2D 类型的实例的列表
plt.legend(lines[:2],['first','second'])
```

```python
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True)
```

```python
# 将图例分不同的区域展示
fig,ax = plt.subplots()
lines=[]
styles= ['-','--','-.',':']
x = np.linspace(0,10,1000)
for i in range(4):
    lines = lines + ax.plot(x,np.sin(x-i*np.pi /2),styles[i],color='black')
ax.axis('equal')

# 设置第一组标签
ax.legend(lines[:2],
          ['line A','line B'],
          loc='upper right',
          frameon = False
         )

# 设置第二组标签
from matplotlib.legend import Legend
leg = Legend(ax,lines[2:],
             ['line C','line D'],
             loc='lower right',
             frameon =False
            )
ax.add_artist(leg)
```

## 多子图

```python
# 在一个10*10的画布中，(0.65,0.65)的位置创建一个0.2*0.2的子图
ax1 = plt.axes()
ax2 = plt.axes([0.65,0.65,0.2,0.2])
```

```python
# 在两个子图中 显示 sin(x) 和cos(x)的图像
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.5,0.8,0.4], ylim=(-1.2,1.2))
ax2 = fig.add_axes([0.1,0.1,0.8,0.4],ylim=(-1.2,1.2))
x = np.linspace(0,10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

```python
# 用for创建6个子图， 并在图中标识对应的子图坐标
for i in range(1,7):
    print(i)
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize=18,ha='center')
```

```python
# 设置同行和列 共享x,y轴
fig,ax = plt.subplots(2,3,sharex='col',sharey='row')
```

```python
# 用[]的方式取出 每个子图， 并添加子图坐标文字
for i in range(2):
    for j in range(3):
        print(i,j)
        ax[i,j].text(0.5,0.5,str((i,j)),fontsize=18, ha='center')
fig
```

```python
# 组合绘制大小不同的子图
grid = plt.GridSpec(2,3,wspace=0.4,hspace=0.3)
plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])
plt.subplot(grid[1,:2])
plt.subplot(grid[1,2])
```

```python
# 显示一组二维数据的频度分布， 并分别在x,y轴上 显示该维度的数据的频度分布
mean = [0,0]
cov= [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,3000).T

fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4,4,hspace=0.2,wspace=0.2)
main_ax = fig.add_subplot(grid[:-1,1:])
y_hist = fig.add_subplot(grid[:-1,0],xticklabels=[],sharey=main_ax)
x_hist = fig.add_subplot(grid[-1,1:],yticklabels=[], sharex=main_ax)

main_ax.scatter(x,y,s=3,alpha=0.2)

x_hist.hist(x,40,histtype='stepfilled',
            orientation = 'vertical'
           )
x_hist.invert_yaxis()

y_hist.hist(y,40,histtype='stepfilled',
           orientation='horizontal'
           )
y_hist.invert_xaxis()
```

## 宝可梦案例

```python
import pandas as pd
df = pd.read_csv('../dataset/matplotlib_data/Pokemon.csv')
```

```python
df.head()
```

```python
plt.style.use('ggplot')
```

```python
# 展示前5个宝可梦的 defense attack hp的堆积条形图
pokemon_name = df['Name'][:5]
pokemon_defense = df['Defense'][:5]
pokemon_attack = df['Attack'][:5]
pokemon_hp = df['HP'][:5]

ind = [x for x,_ in enumerate(pokemon_name)]

plt.figure(figsize=(10,10))
plt.bar(ind,pokemon_defense,width=0.8,label='Defense', color='blue',bottom=pokemon_attack+pokemon_hp)
plt.bar(ind,pokemon_attack,width=0.8,label='Atatck',color='gold',bottom=pokemon_hp)
plt.bar(ind,pokemon_hp,width=0.8,label='Hp', color='red')

plt.xticks(ind,pokemon_name)
plt.ylabel('数值')
plt.xlabel('pokemon')
plt.legend(loc='upper right')
plt.title('5个 pokemon defense & atatck & hp')
```

```python
# 展示前5哥宝可梦的attack ，hp的簇状条形图
N =5
pokemon_hp = df['HP'][:5]
pokemon_attack = df['Attack'][:5]

ind = np.arange(N)
width = 0.35
plt.bar(ind,pokemon_hp,width,label='HP')
plt.bar(ind+width,pokemon_attack,width,label='Attack')

plt.ylabel('Values')
plt.title('宝可梦血量& 攻击')

plt.xticks(ind+width/2, (df['Name'][:5]),rotation=45)
plt.legend(loc='best')
```

```python
# 展示前5个宝可梦的defense,attack ，hp的堆积图
x = df['Name'][:5]
y1 = df['HP'][:5]
y2= df['Attack'][:5]
y3 = df['Defense'][:5]
labels = ["HP","Attack","Defense"]
fig ,ax = plt.subplots()
ax.stackplot(x,y1,y2,y3)
ax.legend(loc='upper left',labels=labels)
plt.xticks(rotation=90)
```

```python
# 公用X轴，展示前5哥宝可梦的defense. Attack ， HP
x = df['Name'][:5]
y1 = df['HP'][:5]
y2= df['Attack'][:5]
y3 = df['Defense'][:5]

fig,(ax1,ax2,ax3) = plt.subplots(3,sharey=True)
ax1.plot(x,y1,'ko-')
ax1.set(title='3 subplots',ylabel='HP')

ax2.plot(x,y2,'r.-')
ax2.set(xlabel='Pokemon', ylabel='Attack')

ax3.plot(x,y3,':')
ax3.set(xlabel='Pokemon', ylabel='Defense')
```

```python
# 展示前15个宝可梦的Atatck, HP 的折线图
plt.plot(df['HP'][:15],'-r',label='HP')
plt.plot(df['Attack'][:15],':g',label='Attack')
plt.legend()
```

```python
# 用scatter的x,y,c属性，展示宝可梦的defense , attack , HP 的数据
x = df['Attack']
y= df['Defense']
colors = df['HP']

plt.scatter(x,y,c =colors, alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.colorbar()
```

```python
# 展示攻击力分布，bins =10
x = df['Attack']
num_bins = 10

n,bins,patches = plt.hist(x,num_bins,facecolor='blue' ,alpha=0.5)
plt.title('Histgram')
plt.xlabel('Attack')
plt.ylabel('Value')
```

```python
# 展示所有宝可梦 type1 的饼图
plt.figure(1,figsize=(8,8))
df['Type 1'].value_counts().plot.pie(autopct="%1.1f%%")
plt.legend()
```

```python
# 展示所有宝可梦 Type 1的柱状图
ax = df['Type 1'].value_counts().plot.bar(figsize=(12,6) ,fontsize=14)
ax.set_title("Pokemon Type 1 Count",fontsize=20)
ax.set_xlabel('Pokemon Type 1',fontsize=20)
ax.set_ylabel('Value' ,fontsize=20)
```

```python
top_10_pokemon = df.sort_values(by='Total', ascending=False).head(10)
corr = top_10_pokemon.corr()
```

```python
# 展示综合评分最高的10只宝可梦的系数间的相关系数矩阵
import seaborn as sns
fig,ax = plt.subplots(figsize=(10,6))
sns.heatmap(corr,annot=True)
ax.set_ylim(9,0)
```
