```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 基本图表
# plot绘制x=(0,10)的sin图像
x = np.linspace(0,10,num=30)
plt.plot(x,np.sin(x))
```

```python
# 用点加线的方式 画出x=(0,10)的 sin图像
plt.plot(x,np.sin(x), '-o')
```

```python
# 用scatter 方法绘制x=(0,10)的 sin图像
plt.scatter(x,np.sin(x))
```

```python
# 使用饼图的面积及颜色绘制一组四维数据
# 伪随机数生成器
rng = np.random.RandomState(0)

x= rng.randn(100)
y= rng.randn(100)
colors = rng.rand(100)
sizes =  1000 * rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # 展示色阶
```

```python
# 绘制一组误差为+- 0.8的数据误差条图
x= np.linspace(0,10,num=50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy, fmt='.k')
```

```python
# 绘制一个柱形图
x= [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]
label = ['A','B','C','D','E','F','G','H']
plt.bar(x,y,tick_label = label)
```

```python
# 绘制一个水平方向的柱形图
# 绘制一个柱形图
x= [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]
label = ['A','B','C','D','E','F','G','H']
plt.barh(x,y,tick_label = label)
```

```python
# 绘制1000哥随机值的直方图
data = np.random.randn(1000)
plt.hist(data)
```

```python
# 设置直方图 分30个bins ，并设置为频率分布
plt.hist(data, bins=30, histtype='stepfilled', density=True)
```

```python
# 在一张图内绘制3组不同的直方图， 并设置透明度
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)

kwargs = dict(alpha=0.3 , bins=40 , density = True)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.show()
```

```python
# 绘制一张二维直方图
mean =[0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,10000).T
plt.hist2d(x,y,bins=30)
plt.show()
```

```python
# 绘制一张 设置网格大小为30 的六角直方图
plt.hexbin(x,y,gridsize=30)
```
