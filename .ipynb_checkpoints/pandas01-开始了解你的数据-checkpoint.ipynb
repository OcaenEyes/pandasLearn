```python
# 被下单数最多的商品是什么
c = chipotle[['item_name','quantity']].groupby(['item_name'],as_index=False).agg({'quantity':sum})
```

```python
# 倒序
c.sort_values(['quantity'],ascending=False)
```

```python
# 共有多少种商品被下单
chipotle['item_name'].nunique()
```

```python
# 购物的选择描述中-下单次数最多的商品是
chipotle['choice_description'].value_counts().head()
```

```python
# 共有多少产品被下
chipotle['quantity'].sum()
```

```python
# 将item_price转换为浮点数
dollarizer = lambda x :float(x[1:-1])
chipotle['item_price'] = chipotle['item_price'].apply(dollarizer)
```

```python
chipotle['item_price']
```

```python
chipotle['sub_total'] = round(chipotle['item_price'] * chipotle['quantity'] ,2)
```

```python
# 累计收入是多少
chipotle['sub_total'].sum()
```

```python
# 共有多少订单
chipotle['order_id'].nunique()
```

```python
# 每一单对应的总价
chipotle[['order_id','sub_total']].groupby(by=['order_id']).agg({'sub_total':'sum'})
```

```python
# 每单订单的平均总价
chipotle[['order_id','sub_total']].groupby(by=['order_id']).agg({'sub_total':'sum'})['sub_total'].mean()
```
