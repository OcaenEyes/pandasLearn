```python
wind2 = pd.read_table(path1,sep='\s+',parse_dates=[[0,1,2]])
wind2
```

```python
# 修复时间bug
def fix_century(x):
    year = x.year -100 if x.year >1989 else x.year
    return datetime.date(year,x.month,x.day)

wind2['Yr_Mo_Dy'] = wind2['Yr_Mo_Dy'].apply(fix_century)
```

```python
wind2.info()
```

```python
wind2.head()
```

```python
# 将日期作为索引，数据类型应该是 datetime64[ns]
wind2['Yr_Mo_Dy'] = pd.to_datetime(wind2['Yr_Mo_Dy'])
```

```python
wind2.info()
```

```python
wind2 = wind2.set_index('Yr_Mo_Dy')
```

```python
wind2
```

```python
# 每个location 共有多少数据缺失值
wind2.isnull().sum()
```

```python
# 对应每个location 共有多少完整的数据值
wind2.shape[0] - wind2.isnull().sum()
```

```python
# 计算风速平均值
wind2.mean()
```

```python
# 对全体数据 计算风速平均值
wind2.mean().mean()
```

```python
# 创建一个loc_stats 的数据框  计算并存储 每个location的风速最小值、最大值、平均值和标准差
loc_stats = pd.DataFrame()
```

```python
loc_stats['min'] = wind2.min()
loc_stats['max'] = wind2.max()
loc_stats['mean'] = wind2.mean()
loc_stats['std'] = wind2.std()
loc_stats
```

```python
# 创建day_stats 以日期汇总计算并存储所有location的风速最小值、最大值、平均值、标准差
day_stats = pd.DataFrame()
```

```python
day_stats['min'] = wind2.min(axis=1)
day_stats['max'] = wind2.max(axis=1)
day_stats['mean'] = wind2.mean(axis=1)
day_stats['std'] = wind2.std(axis=1)
day_stats
```
